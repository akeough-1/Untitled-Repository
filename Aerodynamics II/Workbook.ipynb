```python
"""
Hw 5, P1
A slender missile is flying at Mach 1.5 at low altitude. Assume the wave generated by the nose of the missile is a Mach wave. 
This wave intersects the ground 569 ft behind the nose. At what altitude is the missile flying? 
(Round the final answer to the nearest whole number.)
"""

Mach = 1.5
horz_dist = 569 #ft
tan_mach_angle = (Mach**2 - 1)**(-0.5) #dimensionless
vert_dist = horz_dist*tan_mach_angle
print(f"altitude = {round(vert_dist,0)}ft")
```

```python
"""
Hw 6, P4
Consider a Mach 4 airflow at a pressure of 3 atm. We wish to slow this flow to subsonic speed through a system
of shock waves with as small a loss in total pressure as possible.

Calculate the loss in total pressure for a single normal shock wave. (Round the final answer to one decimal place.)
"""
from Flow_Library import Oblique_Shock, Normal_Shock, Isentropic_Flow
M1 = 4
P1 = 3 #atm

state1 = Isentropic_Flow(M=M1,P1=P1)
P0_1 = state1.P0
shock1 = Normal_Shock("metric",M1=M1,P1=P1,P0_1=P0_1)
P_loss = shock1.P0_2 - P0_1
print(round(P_loss,1))

"""
Calculate the loss in total pressure for an oblique shock with a deflection angle of 25.3°, followed
by a normal shock. (Round the final answer to one decimal place.)
"""
# P0_3 = P0_3/P0_2 * P0_2/P0_1 * P0_1
theta = np.deg2rad(25.3)
shock1 = Oblique_Shock("metric",M1=M1,defl_angle_theta=theta)
shock2 = Normal_Shock("metric",M1=shock1.M2)
P0_3 = shock2.P0_ratio*shock1.P0_ratio*P0_1
P_loss = P0_3 - P0_1
print(round(P_loss,1))

"""
Calculate the loss in total pressure for an oblique shock with a deflection angle of 25.3°, followed
by a second oblique shock of deflection angle of 20°, followed by a normal shock. 
(Round the final answer to one decimal place.)
"""
# P0_4 = P0_4/P0_3 * P0_3/P0_2 * P0_2/P0_1 * P0_1
theta1 = np.deg2rad(25.3)
theta2 = np.deg2rad(20)
shock1 = Oblique_Shock("metric",M1=M1,defl_angle_theta=theta1)
shock2 = Oblique_Shock("metric",M1=shock1.M2,defl_angle_theta=theta2)
shock3 = Normal_Shock("metric",M1=shock2.M2)
P0_4 = shock3.P0_ratio*shock2.P0_ratio*shock1.P0_ratio*P0_1
P_loss = P0_4 - P0_1
print(round(P_loss,1))
```
