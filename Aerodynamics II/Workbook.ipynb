```python
"""
Hw 5, P1
A slender missile is flying at Mach 1.5 at low altitude. Assume the wave generated by the nose of the missile is a Mach wave. 
This wave intersects the ground 569 ft behind the nose. At what altitude is the missile flying? 
(Round the final answer to the nearest whole number.)
"""

Mach = 1.5
horz_dist = 569 #ft
tan_mach_angle = (Mach**2 - 1)**(-0.5) #dimensionless
vert_dist = horz_dist*tan_mach_angle
print(f"altitude = {round(vert_dist,0)}ft")
```

```python
"""
Hw 5, P2
Consider an oblique shock wave with a wave angle of 30° in a Mach 4 flow. The upstream pressure and temperature are 2.65e4 N/m2 and 223.3 K, respectively
(corresponding to a standard altitude of 10,000 m). Calculate the pressure, temperature, Mach number, total pressure, and total temperature behind the wave and 
the entropy increase across the wave. (Round the final answers to two decimal places.)
"""

from Flow_Library import Oblique_Shock as ObS
from Flow_Library import Isentropic_Flow as IsF
from numpy import deg2rad

Mach = 4
beta = deg2rad(30)
P1 = 2.65e4 #Pa
T1 = 223.3 #K
state1 = IsF(M=Mach,P1=P1,T1=T1)

shk = ObS("metric", Mach, wave_angle_beta=beta, P1=P1, T1=T1,
          P0_1=state1.P0, T0_1=state1.T0)
print(f"P2 = {round(shk.P2/1e5,2)}*10^5 Pa")
print(f"T2 = {round(shk.T2,2)} K")
print(f"M2 = {round(shk.M2,2)}")
print(f"P02 = {round(shk.P0_2/1e6,2)}*10^6 Pa")
print(f"T02 = {round(shk.T0_2,2)} K")
print(f"ds = {round(shk.ds,2)} J/kg/K")
```
