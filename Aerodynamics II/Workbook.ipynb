```python
from Flow_Library import Expansion_Fan, Oblique_Shock, Isentropic_Flow
from numpy import deg2rad as d2r, sin, cos, tan

eps1 = d2r(10)
eps2 = d2r(5.0384)
aoa = d2r(0)
M1 = 2.2

# if aoa > eps, 1 is fan, 3 is shock, 2/4 are fan
# if aoa < eps, 1/3 are shock, 2/4 are fan

p1 = Oblique_Shock("metric",M1,defl_angle_theta=aoa+eps1)
p2 = Expansion_Fan("metric",p1.M2,eps1+eps2)
p3 = Oblique_Shock("metric",M1,defl_angle_theta=aoa+eps1)
p4 = Expansion_Fan("metric",p3.M2,eps1+eps2)

c2_ratio = tan(eps1)/(tan(eps2) + tan(eps1))
c1_ratio = 1 - c2_ratio

tau = 2*tan(eps1)*c1_ratio

P1 = p1.P_ratio
P2 = p2.P_ratio*P1
P3 = p3.P_ratio
P4 = p4.P_ratio*P3

fx = (P1 - P2 + P3 - P4)*tau/2
fy = (P3 - P1)*c1_ratio + (P4 - P2)*c2_ratio

d = cos(aoa)*fx + sin(aoa)*fy
l = -sin(aoa)*fx + cos(aoa)*fy

nondim = 2/1.4/M1**2

Cl = nondim*l
Cd = nondim*d

print(Cl, Cd)
```
